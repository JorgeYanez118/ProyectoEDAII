# BFS

```python
class Graph:
    def __init__(self, file_path):
        self.adjacency_list = self.load_graph(file_path)
    
    def load_graph(self, file_path):
        graph = {}
        with open(file_path, 'r') as file:
            for line in file:
                node, neighbors = line.split(":")
                graph[int(node.strip())] = list(map(int, neighbors.split())) if neighbors.strip() else []
        return graph
    
    def bfs(self, start_node):
        """
        Realiza la búsqueda en anchura (BFS) en el grafo desde el nodo inicial.
        :param start_node: Nodo desde donde iniciar la búsqueda.
        :return: Lista del recorrido BFS.
        """
        visited = set()
        queue = [start_node]
        bfs_order = []
        
        while queue:
            current_node = queue.pop(0)
            if current_node not in visited:
                visited.add(current_node)
                bfs_order.append(current_node)
                # Añade a la cola los vecinos no visitados
                queue.extend(neighbor for neighbor in self.adjacency_list[current_node] if neighbor not in visited)
        
        return bfs_order


# Ejemplo de uso
if __name__ == "__main__":
    file_path = "grafo.txt"  # Reemplaza con tu archivo de grafo
    
    # Cargar el grafo y ejecutar BFS
    grafo = Graph(file_path)
    start_node = 1  # Nodo desde donde empezar el BFS
    print("Recorrido BFS:", grafo.bfs(start_node))
```
